1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
x = np.linspace(0, 10, 100)

x_symbol = sp.symbols('x')

f_lin = 10 * x - 5

derivative_f_lin = sp.diff(f_lin, x_symbol)

print("Derivative of 10*x-5 is:", derivative_f_lin)

def fx(x):
  return 10 * x - 5

def fderx(x):
  return x*0 + 10




plt.figure(figsize=(10, 5))
plt.plot(x, fx(x), label='f(x) = 10*x-5', color='blue')
plt.plot(x, fderx(x), label="f'(x) = 10", color='green')
```

Since derivative of 10x-5 as of any linear finction is just a number, than we can see that derivative is always 10, and the change of the function is constant. So amount of hours affets mark linearly thus for each hour spent mark rises by 10.
